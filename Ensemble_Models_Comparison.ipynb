# Part 1: Data Preparation

1. Import necessary libaries.

```python
import pandas as pd
```

2. Read in the data and view the first 10 rows.

```python
cancer_data = pd.read_csv("data.csv", sep=",", skipinitialspace=True)
cancer_data.head(10)
```

3. Clean data.
- Remove id column.
- Remove last empty column.

```python
cancer_data = cancer_data.drop(['id', 'Unnamed: 32'], axis=1)
```

4. View the data again to make sure the columns were deleted.

```python
cancer_data.head(10)
```

5. Split dataset into two parts. X for attributes and y for lables.

```python
X = cancer_data.drop('diagnosis', axis=1)
y = cancer_data['diagnosis']
```

6. Convert the class lables to numerical values so they can be used for score functions. 
- B = 0
- M = 1

```python
y_numerical = y.replace(['B', 'M'], [0, 1])
```

# Part 2: Model Creation and Evaluation

1. Import necessary libaries and define a variable for the random state.

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_validate
RANDOM_STATE = 1111
```

## Part A: Random Forest

1. Create the Random Forest classfier.

```python
rf = RandomForestClassifier(n_estimators=100, random_state = RANDOM_STATE)
```

2. Perform a 5 fold cross-validation and get the score metrics.

```python
rf_scores = cross_validate(rf, X, y_numerical, cv=5, scoring=('recall', 'precision', 'f1', 'accuracy', 'roc_auc', 'average_precision'))
```

## Part B: AdaBoost

1. Create the AdaBoost classifier.

```python
ab = AdaBoostClassifier(n_estimators=100, random_state = RANDOM_STATE)
```

2. Perform a 5 fold cross-validation and get the score metrics.

```python
ab_scores = cross_validate(ab, X, y_numerical, cv=5, scoring=('recall', 'precision', 'f1', 'accuracy', 'roc_auc', 'average_precision'))
```

## Part 4: Compile Scores

1. Compile the scores and the differences between the two classifiers into a dataframe and display them.

```python
indexes = []
rf_mean_scores = []
for key, value in rf_scores.items():
    indexes.append(key)
    rf_mean_scores.append(np.mean(value))

ab_mean_scores = []
for key, value in ab_scores.items():
    ab_mean_scores.append(np.mean(value))

differences = np.subtract(ab_mean_scores, rf_mean_scores)

data = {'Random Forest':rf_mean_scores, 'AdaBoost': ab_mean_scores, 'AdaBoost - Random Forest': differences}
metrics = pd.DataFrame(data, index = indexes)
metrics
```
